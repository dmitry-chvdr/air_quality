```python
import pandas as pd
import glob
import os
import matplotlib.pyplot as plt
from methods import get_result_data, make_processed_dataframes
%matplotlib inline 
```

```python
main_path = "data/"
dirs = ["CO", "NO2", "O3", "PM10", "PM2.5", "SO2"]
```

```python
def draw_plot(dataframes, name):
    data = dataframes[name]
    plt.figure(figsize = (25, 5))
    plt.title(f"Time Series {name}")
    plt.plot(data["date"], data["Concentration"], label="Day mean")
    plt.legend(loc="upper left")
    plt.xlabel('Year')
    plt.show()
```

```python
processed_dfs = make_processed_dataframes(main_path, dirs)
```

По временным рядам SO2, PM2.5, PM10.Самые ярко выраженные выбросы приходятся на конец года, что может быть связано с использованием большого количества пиротехники. По NO2 можно заметить выбросы из года в год в один и тот же период.
Также наблюдается некая сезонность по временным рядам CO, PM2.5, PM10 - в холодное время года концентрация увеличивется.
По О3 также можно заметить увелечение концентрации в тёплое время года.

```python
aqi_data = get_result_data(main_path, dirs, False)
```

```python
plt.figure(figsize = (25, 5))
plt.title(f"Time Series AQI")
plt.plot(aqi_data["Date"], aqi_data["aqi_index"], label="Day")
plt.legend(loc="upper left")
plt.xlabel('Year')
plt.show()
```

```python
plt.figure(figsize = (25, 5))
plt.title(f"Time Series AQI 2 Years")
plt.plot(aqi_data["Date"][:730], aqi_data["aqi_index"][:730], label="Day")
plt.legend(loc="upper left")
plt.xlabel('Year')
plt.show()
```

```python
plt.figure(figsize = (25, 5))
plt.title(f"Time Series AQI Year")
plt.plot(aqi_data["Date"][:365], aqi_data["aqi_index"][:365], label="Day")
plt.legend(loc="upper left")
plt.xlabel('Year')
plt.show()
```

```python
Можно заметить сезонность на графика, что в теплое время года индекс увеличивается.
```
